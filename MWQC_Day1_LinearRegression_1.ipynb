---
While we may not be able to find the exact $\theta_{0}$ and $\theta_{1}$ values, our objective is to find an optimum set of parameters $\hat{\theta}_{0}$ and $\hat{\theta}_{1}$ such that the predicted y values (lets call them $\hat{y}$:

$$\hat{y}=\hat{\theta}_{0}+\hat{\theta}_{1}x$$
---
fits the original data as closely as possible.

Least Squares Fit
---

The simplest way to optimize the parameters $\theta_{0}$ and $\theta_{1}$ is to minimize the Mean Squared Error (MSE) between the actual $y$ values and the predicted y values ($\hat{y}$). This is called a Least Squared Fitting and gives us the following analytical expressions for the parameters:

$\hat{\theta}_1 = \frac{\sum_{i}^{n} (x_i-\overline{x})(y_i - \overline{y})}{\sum_{i}^n(x_i - \overline{x})^2 }$ or  $\hat{\theta}_1 = \frac{\overline{xy} - (\overline{x})(\overline{y})} {\overline{x^2} - (\overline{x})^2}$ <br> $\hat{\theta}_0 = \overline{y} - \hat{\theta}_1 \overline{x}$  or  $\hat{\theta}_0 = \frac{(\overline{x^2})(\overline{y}) - (\overline{x})(\overline{xy})}{\overline{x^2} - (\overline{x})^2}$<br>

Write the code to perform linear regression using least squares fitting. You may follow the instructions given below:
* Define a function `fit_leastsq(x,y)` that returns the fit parameters `theta0_hat` and `theta1_hat` using the formulae given above.
* Pass the `x` and `y` data in the Anscombe quartet's dataset I as arguments to the `fit_leastsq()` function. Store the optimized parameters returned by the function in suitable variables and print their values.
* Generate a set of artificial 100 `x` values and predict the corresponding `y_hat` values using the relation $\hat{y}=\hat{\theta}_{0}+\hat{\theta}_{1}x$.
* Plot the predicted `y_hat` values as a line and the original `y` values as points to see the linear fit.
* Similarly, fit a line for the data in `df_III`.  

```python
#Write your code here
def fit_leastsq(x,y):
    xav = np.mean(x)
    yav = np.mean(y)

    theta1_hat = (np.mean(x*y)-(xav*yav))/(np.mean(x*x)-(xav*xav))
    theta0_hat = ((np.mean(x*x)*np.mean(y))-(np.mean(x)*np.mean(x*y)))/(np.mean(x*x)-(xav*xav))
    return(theta0_hat,theta1_hat)
```

One easy way is to measure the Coefficient of Determination $R^{2}$ (also called Goodness of Fit) defined as the ratio of the SSR (sum of squared residuals) to the TSS (total sum of squares):
$$
R^{2}=1-\frac{SSR}{TSS}  = 1 - \frac{\sum_i \left(y - \hat{y}_i\right)^2}{\sum_i \left(y - \bar{y}\right)^2}
$$
$R^{2}$ is unitless, it's $1$ when the fit is perfect, and $0$ when the fit is awful. Note that in the case of linear regression, $R$ is the same as the correlation coefficient.

**Michaelis-Menten Enzyme Kinetics:**

The rate at which an enzyme works is crucial for understanding many biological processes. The general reaction scheme of an enzyme-catalyzed reaction is as follows:
$$
E+S \xrightarrow[k_{-1}]{k_{1}} ES \xrightarrow {k_{2}} E + P
$$
Here the enzyme $E$ binds to the substrate $S$ with a forward rate constant $k_{1}$ to form the enzyme-substrate complex $ES$. The $ES$ complex can form the product $P$ and release the enzyme with a rate constant $k_{2}$. Or it can fall apart before successful catalysis with a rate constant $k_{-1}$.

The Michaelis-Menten equation describes how the reaction rate ($v$) of an enzyme-catalyzed reaction depends on the concentration of the substrate ($S$).
$$
v = \frac{V_{max}[S]}{K_{m}+[S]}
$$
Here $K_{m}$ is called the Michaelis constant defined as $(k_{-1}+k_{2})/k_{1}$ and $V_{max}$ is the maximum rate of the reaction. We can evaluate these two parameters from experimental measurements of the rate at a given substrate concentration by taking a reciprocal of the Michaelis-Menten equation.
$$
\frac{1}{v}=\frac{K_{m}}{V_{max}[S]}+\frac{1}{V_{max}}
$$
Clearly a plot of $1/v$ vs $1/[S]$ will be a straight line; with slope $K_{m}/V_{max}$ and intercept $1/V_{max}$.

Multiple Linear Regression (or Multivariate Linear Regression)
---

So far, we have seen linear regression in one variable. But what if the data is linearly dependent on more than one variable. Multiple Linear Regression is an extension of simple linear regression where we use multiple independent variables to predict the dependent variable. This allows us to model more complex relationships and improve the predictive power of our regression model.

The multivariate linear regression model can be represented by the equation:
$$
y=\theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2} + \theta_{3}x_{3} + ... + \theta_{D-1}x_{D-1}
$$
where $x_{1},x_{2},...,x_{n}$ are the independent variables. We can write this in matrix notation as:
$$\mathbf{Y = X^{T}.\theta}$$

A Least Squares Analysis can be applied here as well, yielding the following analytical solution for the optimized parameters:

$$
\mathbf{\hat{\theta}=(X^{T}X)^{-1}X^{T}Y}
$$

Gradient Descent
---

Optimizing a general Objective Funtion $J(\theta)$ (also called Cost Function or Loss Function) effectively amounts to a gradient descent in the space of the parameters.The parameters are updated every step based on the value of the gradient (of $J(\theta)$) such that:
$$
\theta_{i} := \theta_{i} - \alpha \frac{\partial J(\theta)}{\partial \theta_{i}}
$$
where $\alpha$ is a hyper-parameter called the Learning Rate which determines how fast the parameters are updated. It should be chosen very carefully. We keep upgrading the parameters until the the objective function can be minimized no further i.e. the difference between successive values of the objective function is negligible.

In the case of Linear Regression, this is equivalent to optimizing the Mean Squared Error (MSE) i.e. $J(\theta)=\frac{1}{2} \sum_{i=1}^{N}(y_{i}-\hat{y}_{i})^{2}$ in the space of the coefficients $\hat{\theta}_{0}$ and $\hat{\theta}_{1}$. We know that the predicted $\hat{y}$ values are given by a hypothesis $h_{\theta}(x)=\theta_{0}+\theta_{1}x$. So the update equations become:
$$
\theta_{0} := \theta_{0} - \alpha \frac{\partial J(\theta)}{\partial \theta_{0}} \quad \text{or} \quad \theta_{0} :=  \theta_{0} - \alpha \sum_{i=1}^{N}[h_{\theta}(x_{i})-y_{i}]
$$
$$
\theta_{1} := \theta_{1} - \alpha \frac{\partial J(\theta)}{\partial \theta_{1}} \quad \text{or} \quad \theta_{1} :=  \theta_{1} - \alpha \sum_{i=1}^{N}x_{i}[h_{\theta}(x_{i})-y_{i}]
$$
The coefficients are updated iteratively until the loss becomes less than a tolerance value (another hyperparameter). Each such iteration is often called an **Epoch** in Machine Learning parlance.

**Standard Errors**

We assume the slope and intercept to follow a Student's T distribution and calculate the Standard Error (S.E) in the slope and intercept respectively as:

$$
S.E_{\theta_{1}}=\frac{\sigma}{\sqrt{\sum_{i=1}^{n}(x_{i}-\bar{x})^{2}}}
$$

$$
S.E_{\theta_{0}}=\sigma\sqrt{\left[\frac{1}{n}+\frac{\bar{x}^{2}}{\sum_{i=1}^{n}(x_{i}-\bar{x})^{2}}\right]}
$$
where
* $\sigma$ is the standard deviation equal to $\sqrt{\frac{\sum_{i=1}^{n}(y_{i}-\hat{y}_{i})^{2}}{df}}$. Here df=n-2
* n is the number of samples in the data.

**Hypothesis Testing and p-value:**
How do we know that there should be a slope and intercept at all? What if the x and true y values are not correlated at all? In such a scenario, we we first frame a **null-hypothesis** and measure the **p-value** to validate it.


**Testing for the intercept**

First let us deal with the intercept. In this case, our null hypothesis ($H_{o}$) is:

$H_{o}$: There is no intercept i.e. $\theta_{0}=0$.

Associated with every null hypothesis is an alternate hypothesis $H_{a}$ which in this case is:

$H_{a}: \theta_{0} \neq 0$

Note that we usually choose a statement as Null hypothesis when we want it to be false.

So we can reject the Null hypothesis only when the p-value is less than the significance level $\alpha$. The p-value is calculated as:

$$
p=2(1-CDF_{t}(|T_{\theta_{0}}|))
$$
where $CDF_{t}$ is the cumulative distribution function of the T-distribution with **n-1** degrees of freedom. This is because we work under the assumption of the null hpyothesis (i.e., only slope needs to be fit)
